```python
import pandas as pd
import sys
import os
import re

def clean_ptid(ptid_value):
    """
    Clean ptid by removing 'P-' prefix and leading zeros, keeping everything after zeros.
    
    Examples:
    P-0001 -> 1
    P-0123 -> 123
    P-1234 -> 1234
    P-00123a -> 123a
    P-0056b -> 56b
    
    Parameters:
    ptid_value: The ptid value to clean
    
    Returns:
    str or None: The cleaned value after P- and leading zeros, or None if invalid format
    """
    try:
        # Convert to string in case it's read as a number
        ptid_str = str(ptid_value).strip()
        
        # Check if it starts with P-
        if ptid_str.startswith('P-'):
            # Extract everything after P-
            after_prefix = ptid_str[2:]  # Remove 'P-'
            
            # Remove leading zeros using lstrip
            cleaned_value = after_prefix.lstrip('0')
            
            # If all characters were zeros, keep one zero
            if not cleaned_value:
                cleaned_value = '0'
            
            return cleaned_value
        else:
            print(f"Warning: ptid doesn't start with 'P-': {ptid_str}")
            return None
            
    except (ValueError, AttributeError):
        print(f"Warning: Could not process ptid: {ptid_value}")
        return None

def filter_excel_by_ptid_pattern(file1_path, file2_path, output_path):
    """
    Reads ptid from file1, cleans the pattern (removes P- and leading zeros),
    and filters file2 to keep only rows with matching Patient Id values.
    
    Parameters:
    file1_path (str): Path to Excel file containing 'ptid' column with P-**** pattern
    file2_path (str): Path to Excel file containing 'Patient Id' column (file to be filtered)
    output_path (str): Path where the filtered file will be saved
    """
    
    try:
        # Read the first Excel file to get ptid values
        print(f"Reading ptid values from: {file1_path}")
        df1 = pd.read_excel(file1_path)
        
        # Check if 'ptid' column exists in file1
        if 'ptid' not in df1.columns:
            print(f"Error: Column 'ptid' not found in {file1_path}")
            print(f"Available columns in file1: {list(df1.columns)}")
            return False
        
        print(f"File1 has {len(df1)} rows with ptid column")
        
        # Clean ptid values - remove P- prefix and leading zeros
        print("Processing ptid values (removing P- prefix and leading zeros)...")
        cleaned_ptids = []
        ptid_order = []  # Keep track of order from file1
        
        for idx, ptid in enumerate(df1['ptid'].dropna()):
            cleaned = clean_ptid(ptid)
            if cleaned is not None:
                cleaned_ptids.append(cleaned)
                ptid_order.append((cleaned, idx))  # Store cleaned value with original order
        
        # Create ordered list maintaining file1 sequence
        ordered_ptids = [item[0] for item in sorted(ptid_order, key=lambda x: x[1])]
        
        # Convert to set for faster lookup (keep as strings to handle alphanumeric)
        reference_patient_ids = set(str(x) for x in cleaned_ptids)
        print(f"Successfully processed {len(reference_patient_ids)} unique cleaned patient IDs")
        
        # Show some examples of the conversion
        sample_original = list(df1['ptid'].dropna().head(5))
        sample_cleaned = [clean_ptid(x) for x in sample_original]
        print(f"Example conversions:")
        for orig, clean in zip(sample_original, sample_cleaned):
            if clean is not None:
                print(f"  {orig} -> {clean}")
        
        # Read the second Excel file to be filtered
        print(f"\nReading target file: {file2_path}")
        df2 = pd.read_excel(file2_path)
        
        # Check if 'Patient Id' column exists in file2
        if 'Patient Id' not in df2.columns:
            print(f"Error: Column 'Patient Id' not found in {file2_path}")
            print(f"Available columns in file2: {list(df2.columns)}")
            return False
        
        print(f"File2 has {len(df2)} total rows before filtering")
        
        # Filter file2 to keep only rows where Patient Id matches cleaned ptid values
        # Convert Patient Id to string for comparison to handle mixed data types
        df2_patient_ids_str = df2['Patient Id'].astype(str)
        
        # Find rows that will be removed (Patient IDs not in reference list)
        rows_to_remove = ~df2_patient_ids_str.isin(reference_patient_ids)
        removed_patient_ids = df2_patient_ids_str[rows_to_remove].unique()
        
        # Print the Patient IDs that are being removed
        print(f"\nPatient IDs being removed ({len(removed_patient_ids)} unique IDs):")
        if len(removed_patient_ids) <= 50:  # Show all if 50 or fewer
            print(f"Removed Patient IDs: {sorted(removed_patient_ids.tolist())}")
        else:  # Show first 50 if more than 50
            print(f"First 50 removed Patient IDs: {sorted(removed_patient_ids.tolist()[:50])}")
            print(f"... and {len(removed_patient_ids) - 50} more")
        
        # Perform the actual filtering
        filtered_df = df2[df2_patient_ids_str.isin(reference_patient_ids)]
        
        # Reorder the filtered dataframe to match the ptid order from file1
        print("Reordering output to match ptid sequence from file1...")
        
        # Create a mapping of Patient Id to order based on file1 ptid sequence
        patient_id_to_order = {}
        for order_idx, ptid_value in enumerate(ordered_ptids):
            if ptid_value not in patient_id_to_order:  # Handle duplicates by keeping first occurrence
                patient_id_to_order[ptid_value] = order_idx
        
        # Add order column for sorting
        filtered_df_copy = filtered_df.copy()
        filtered_df_copy['_temp_order'] = filtered_df_copy['Patient Id'].astype(str).map(patient_id_to_order)
        
        # Sort by the order from file1, then drop the temporary order column
        ordered_filtered_df = filtered_df_copy.sort_values('_temp_order').drop('_temp_order', axis=1)
        
        print(f"\nAfter filtering and reordering: {len(ordered_filtered_df)} rows remain")
        print(f"Removed {len(df2) - len(ordered_filtered_df)} rows")
        
        # Save the filtered and ordered dataframe to output file
        ordered_filtered_df.to_excel(output_path, index=False)
        print(f"Filtered and ordered data saved to: {output_path}")
        
        # Display detailed summary
        print("\n" + "="*60)
        print("DETAILED FILTERING SUMMARY")
        print("="*60)
        print(f"Reference file: {file1_path}")
        print(f"  - Original ptid values (P-*** format): {len(df1['ptid'].dropna())}")
        print(f"  - Successfully cleaned IDs: {len(reference_patient_ids)}")
        
        print(f"\nTarget file: {file2_path}")
        print(f"  - Original rows: {len(df2)}")
        print(f"  - Rows kept (matching Patient Id): {len(ordered_filtered_df)}")
        print(f"  - Rows removed: {len(df2) - len(ordered_filtered_df)}")
        print(f"  - Percentage kept: {len(ordered_filtered_df)/len(df2)*100:.1f}%")
        print(f"  - Output ordered by file1 ptid sequence: ✓")
        
        print(f"\nOutput: {output_path}")
        
        # Show matching analysis
        file2_patient_ids = set(str(x) for x in df2['Patient Id'].dropna())
        matched_ids = reference_patient_ids.intersection(file2_patient_ids)
        unmatched_ids = reference_patient_ids - file2_patient_ids
        
        print(f"\nID Matching Analysis:")
        print(f"  - Cleaned ptids that found matches: {len(matched_ids)}")
        print(f"  - Cleaned ptids with no matches: {len(unmatched_ids)}")
        
        if len(unmatched_ids) <= 15 and unmatched_ids:
            print(f"  - Unmatched cleaned IDs: {sorted(list(unmatched_ids))}")
        
        return True
        
    except FileNotFoundError as e:
        print(f"Error: Could not find file - {e}")
        return False
    except PermissionError as e:
        print(f"Error: Permission denied - {e}")
        print("Make sure the output file is not open in Excel")
        return False
    except Exception as e:
        print(f"Unexpected error: {e}")
        return False

def preview_ptid_patterns(file1_path):
    """
    Preview ptid patterns in file1 to verify they match expected P-**** format
    """
    try:
        print("PREVIEWING PTID PATTERNS")
        print("="*40)
        
        df1 = pd.read_excel(file1_path)
        
        if 'ptid' not in df1.columns:
            print(f"Error: 'ptid' column not found")
            return
        
        # Show sample ptid values
        ptid_values = df1['ptid'].dropna().head(10)
        print(f"Sample ptid values from {file1_path}:")
        
        for ptid in ptid_values:
            cleaned = clean_ptid(ptid)
            status = "✓" if cleaned is not None else "✗"
            print(f"  {status} {ptid} -> {cleaned}")
        
        # Check for any non-standard patterns
        all_ptids = df1['ptid'].dropna()
        pattern_matches = sum(1 for ptid in all_ptids if clean_ptid(ptid) is not None)
        
        print(f"\nPattern Analysis:")
        print(f"  - Total ptid values: {len(all_ptids)}")
        print(f"  - Valid P-**** patterns: {pattern_matches}")
        print(f"  - Invalid patterns: {len(all_ptids) - pattern_matches}")
        
    except Exception as e:
        print(f"Error previewing file: {e}")

def main():
    """
    Main function - UPDATE THE FILE PATHS BELOW
    """
    
    # UPDATE THESE PATHS TO MATCH YOUR ACTUAL FILES
    file1_path = "Final_clinical_data_793_patients.xlsx"    # File with 'ptid' column (P-**** format)
    file2_path = "821 demographics 20241204 for NAS.xlsx"    # File with 'Patient Id' column (to be filtered)
    output_path = "filtered_output.xlsx"  # Your desired output file name
    
    print("PTID Pattern Matcher and Excel Filter")
    print("="*40)
    print("This program will:")
    print("1. Read ptid values from file1 (P-*** format, any length)")
    print("2. Convert P-0001 -> 1, P-0123 -> 123, P-00123a -> 123a, etc.")
    print("3. Keep only rows in file2 where 'Patient Id' matches these cleaned values")
    print()
    print(f"File1 (ptid source): {file1_path}")
    print(f"File2 (to filter): {file2_path}")
    print(f"Output file: {output_path}")
    print()
    
    # Validate input files exist
    if not validate_files(file1_path, file2_path):
        print("Please update the file paths in the main() function and try again.")
        return
    
    # Optional: Preview ptid patterns first
    print("Previewing ptid patterns...")
    preview_ptid_patterns(file1_path)
    print()
    
    # Run the filtering process
    success = filter_excel_by_ptid_pattern(file1_path, file2_path, output_path)
    
    if success:
        print("\n" + "="*50)
        print("✓ PROCESSING COMPLETED SUCCESSFULLY!")
        print("="*50)
        print(f"Your filtered results are saved in: {output_path}")
    else:
        print("\n" + "="*50)
        print("✗ PROCESSING FAILED")
        print("="*50)
        print("Please check the error messages above and fix any issues.")

if __name__ == "__main__":
    main()
```

```python
import os

def get_folder_names_csv(directory_path):
    """
    Reads all folder names from the given directory path and returns them as a comma-separated string.
    """
    if not os.path.exists(directory_path):
        raise FileNotFoundError(f"Path does not exist: {directory_path}")

    # List only folders (exclude files)
    folder_names = [f for f in os.listdir(directory_path) if os.path.isdir(os.path.join(directory_path, f))]

    # Join as comma-separated string
    folder_names_csv = ', '.join(folder_names)

    return folder_names_csv

# Example usage
if __name__ == "__main__":
    path = r"E:/AE_TILE_EXTRACTION/4.NEW_TRAINING_DATA/TRAINING_SAMPLES/STANFORD"  # Change to your directory path
    try:
        result = get_folder_names_csv(path)
        print("Folder names (CSV):")
        print(result)
    except Exception as e:
        print(f"Error: {e}")  
```
